##Redes Neuronales - Clasificacion binaria
En este caso, se necesita de un modelo de clasificacion binaria debido a que la varible independiente solo tiene dos posibles valores, 0 (El cliente no abandono) y 1 (el cliente abandono).

```python
print(y_trainNp.shape)
y_trainNp = np.squeeze(y_trainNp)
```

Establecemos la cantidad de capas que tendra nuestra red neuronal. En este caso solo tiene 3:
* La capa de entrada para las 94 caracteristicas
* Una capa oculta con 10 unidades
* La capa de salida para dos etiquetas (0 y 1).

```python
# Configurando parametros necesario
input_layer_size  = 94  # Entrada de 94 caracteristicas
hidden_layer_size = 10  # 10 unidades ocultas
num_labels = 2          # 2 etiquetas, 0 y 1
```

Definimos la funcion **sigmoid** que es la funcion de activacion que convierte los valores de entrada en valores entre 0 y 1.

Tambien definimos la funcion **sigmoidGradient** que calcula el gradiente de la funcion sigmoide. Esta funcion es utilizada en el **Backpropagation** para actualizar los pesos de la red.

```python
def sigmoid(z):

    return 1.0 / (1.0 + np.exp(-z))


def sigmoidGradient(z):

    g = np.zeros(z.shape)

    g = sigmoid(z) * (1 - sigmoid(z))

    return g
```

Definimos la funcion **nnCostFunction** que calculara la funcion de costo y el gradiente de la red neuronal.

```python
def nnCostFunction(nn_params,
                   input_layer_size,
                   hidden_layer_size,
                   num_labels,
                   X, y, lambda_= 0.0):

    #reconstruyendo las matrices de pesos
    Theta1 = np.reshape(nn_params[:hidden_layer_size * (input_layer_size + 1)],
                        (hidden_layer_size, (input_layer_size + 1)))

    Theta2 = np.reshape(nn_params[(hidden_layer_size * (input_layer_size + 1)):],
                        (num_labels, (hidden_layer_size + 1)))

    m = y.size #cantidad de ejemplos del conjunto de entrenamiento
    #inicializando la funcion de costo y los gradientes de las matrices de pesos
    J = 0
    Theta1_grad = np.zeros(Theta1.shape)
    Theta2_grad = np.zeros(Theta2.shape)

    # Forward
    a1 = np.concatenate([np.ones((m, 1)), X], axis=1) #adicionando el termino de sesgo

    a2 = sigmoid(a1.dot(Theta1.T))
    a2 = np.concatenate([np.ones((a2.shape[0], 1)), a2], axis=1)

    a3 = sigmoid(a2.dot(Theta2.T))

    #convierte y en una matriz binaria
    y_matrix = y.reshape(-1)
    y_matrix = np.eye(num_labels)[y_matrix]

    #copiando los pesos
    temp1 = Theta1
    temp2 = Theta2

    #calcula el termino de regularizacion
    reg_term = (lambda_ / (2 * m)) * (np.sum(np.square(temp1[:, 1:])) + np.sum(np.square(temp2[:, 1:])))

    # función de costo logística binaria con el termino de regularizacion
    J = (-1 / m) * np.sum((np.log(a3) * y_matrix) + np.log(1 - a3) * (1 - y_matrix)) + reg_term

    # Backpropagation
    delta_3 = a3 - y_matrix #se calcula el error de las predicciones
    delta_2 = delta_3.dot(Theta2)[:, 1:] * sigmoidGradient(a1.dot(Theta1.T)) #se propaga el error hacia atras
    # calculando los gradientes acumulados de los pesos.
    Delta1 = delta_2.T.dot(a1)
    Delta2 = delta_3.T.dot(a2)
    #se calcula los gradientes promedio de Theta1 y se adiciona el termino de regularizacion
    Theta1_grad = (1 / m) * Delta1
    Theta1_grad[:, 1:] = Theta1_grad[:, 1:] + (lambda_ / m) * Theta1[:, 1:]
    #se calcula los gradientes promedio de Theta2 y se adiciona el termino de regularizacion
    Theta2_grad = (1 / m) * Delta2
    Theta2_grad[:, 1:] = Theta2_grad[:, 1:] + (lambda_ / m) * Theta2[:, 1:]
    #pone todos los gradientes en un solo vector
    grad = np.concatenate([Theta1_grad.ravel(), Theta2_grad.ravel()])

    return J, grad
```

Definimos la funcion **randInitializeWeights** para inicializar los pesos de la red neuronal de forma aleatoria en un rango pequeño.

```python
def randInitializeWeights(L_in, L_out, epsilon_init=0.12):

    W = np.zeros((L_out, 1 + L_in))
    #generan pesos aleatorios en el rango -epsilon_init y epsilon_init
    W = np.random.rand(L_out, 1 + L_in) * 2 * epsilon_init - epsilon_init

    return W
```

Inicializamos los pesos y los unimos en un solo vector.

```python
print('Inicialización de parámetros de redes neuronales...')

initial_Theta1 = randInitializeWeights(input_layer_size, hidden_layer_size)
initial_Theta2 = randInitializeWeights(hidden_layer_size, num_labels)

# Desenrrollar parametros
initial_nn_params = np.concatenate([initial_Theta1.ravel(), initial_Theta2.ravel()], axis=0)
```

Definimos la funcion **featureNormalize** para normalizar las caracteristicas. Esto convierte los valores para que esten en el rango de -1 y 1. La normalizacion se realiza para que el aprendizaje del modelo converja mas rapido al entrenar, ademas en este caso era completamente necesario ya que el entrenamiento del modelo con los datos sin normalizar causaba desbordamiento.

```python
def  featureNormalize(X):
    X_norm = X.copy()
    mu = np.zeros(X.shape[1])
    sigma = np.zeros(X.shape[1])

    mu = np.mean(X, axis = 0)#calcula la media de cada columna
    sigma = np.std(X, axis = 0)#calcula la desviacion estandar de cada columna
    X_norm = (X - mu) / sigma #normaliza la matriz X (-1, 1)

    return X_norm
```

Normalizamos las caracteristicas

```python
# llama featureNormalize con los datos cargados
X_NormtrainNp = featureNormalize(X_trainNp)

print(X_trainNp)
print(X_NormtrainNp)
```

###Entrenamiento de la red neuronal
Para el entrenamiento se utilizo la funcion **minimize** del modulo **optimize** de **scipy** para encontrar los parametros que minimizan la funcion de costo. Se utilizo el metodo **L-BFGS-B** ya que es adecuado para casos en los que existe un gran numero de variables, ademas de que permite definir un numero maximo de iteraciones.

De esta manera optenemos los pesos aprendidos para cada conexion entre capas de la red neuronal.

```python
#lambda se define al igual que el numero de veces que se optimizará
lambda_ = 0.1
max_iterations = 20000

# La funcion de costo es aplicada al modelo
costFunction = lambda p: nnCostFunction(p, input_layer_size, hidden_layer_size, num_labels, X_NormtrainNp, y_trainNp, lambda_)

#funcion que permite minimizar la funcion de costo
res = optimize.minimize(costFunction, initial_nn_params, jac=True,
                        method='L-BFGS-B',  # Cambio de metodo 'L-BFGS-B', para poder definir las iteraciones
                        options={'maxiter': max_iterations})
# obteniendo el resultado de la optimizacion
nn_params = res.x

# Obtenemos Theta1 y Theta2 a partir de nn_params
Theta1 = np.reshape(nn_params[:hidden_layer_size * (input_layer_size + 1)],
                    (hidden_layer_size, (input_layer_size + 1)))

Theta2 = np.reshape(nn_params[(hidden_layer_size * (input_layer_size + 1)):],
                    (num_labels, (hidden_layer_size + 1)))
```

Definimos la funcion **predict** que recibe los pesos aprendidos en el entrenamiento de la red neuronal y los datos de prueba para realizar las prediciones.

```python
def predict(Theta1, Theta2, X):

    m = X.shape[0] #numero de ejemplos
    num_labels = Theta2.shape[0] #numero de etiquetas

    p = np.zeros(m)
    #calcula las activaciones de la capa oculta
    h1 = sigmoid(np.dot(np.concatenate([np.ones((m, 1)), X], axis=1), Theta1.T))
    #calcula las activaciones para la capa de salida
    h2 = sigmoid(np.dot(np.concatenate([np.ones((m, 1)), h1], axis=1), Theta2.T))
    p = np.argmax(h2, axis=1) #encuentra el indice del valor maximo para cada ejemplo
    return p
```
